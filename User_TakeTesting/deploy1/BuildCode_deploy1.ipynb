````python
import pandas as pd

# Load the Excel file and the specific sheet
file_path = r'D:\OneDrive - Hanoi University of Science and Technology\ITE10-DS&AI-HUST\Learn&Task\PRODUCT_THECOACH\TASK5_GPTschatbot_TestingSale\Input\[SaleViewer]Define_b√†i_to√°n_SalerTesting.xlsx'
sheet_name = 'TN_P1P2'
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Extract relevant columns
questions = df['question']
checking_answers = df['checking_answer']
explain_answers = df['explain_answer']

# Initialize score
total_score = 0

# Iterate through each question
for i in range(len(questions)):
    # Check if the question is None
    if pd.isna(questions[i]):
        break
    
    # Display the question
    print(f"\nC√¢u h·ªèi: {questions[i]}")
    
    # Prompt the user to enter their answer
    user_answer = input("Nh·∫≠p c√¢u tr·∫£ l·ªùi c·ªßa b·∫°n (A/B/C/D): ").strip().upper()
    
    # Compare user answer with the correct answer
    if user_answer == checking_answers[i]:
        assistant_response = f"ƒê√∫ng! {explain_answers[i]}"
        point = 1
        total_score += 1
    else:
        assistant_response = f"Sai! {explain_answers[i]}"
        point = 0
    
    # Display the result
    print(f"user_answer: {user_answer}")
    print(f"point: {point}")
    print(f"assistant_response: {assistant_response}")


# Display total score
print(f"\nT·ªïng ƒëi·ªÉm c·ªßa b·∫°n l√†: {total_score}/{len(questions)}")
````

#### 1.1 UI Tr·∫Øc Nghi·ªám

Build UI with Gradio 
1. Giao di·ªán b·∫Øt ƒë·∫ßu: c√≥ n√∫t `Start`
2. Giao di·ªán khi ƒë√£ nh·∫•n `Start`:
    - √î `response_and_question`: 
        - Ban ƒë·∫ßu c√¢u h·ªèi 1 th√¨ `response` r·ªóng => UI hi·ªÉn th·ªã `response_and_question` l√† c√¢u h·ªèi 1
        - Sau khi ·∫•n Enter th√¨ m√†n h√¨nh hi·ªÉn th·ªã √¥ `response_and_question` l√†: response c·ªßa c√¢u 1 v√† question c√¢u 2 v√† √¥ `user_input` ƒë·ªÉ input c√¢u tr·∫£ l·ªùi c·ªßa user
        - User nh·∫≠p ƒë√°p √°n v√†o √¥ `user_input` v√† ·∫•n Enter => UI hi·ªÉn th·ªã √¥ `response_and_question` l√† response c√¢u 2 v√† question c√¢u 3 v√† √¥ `user_input` ƒë·ªÉ input c√¢u tr·∫£ l·ªùi c·ªßa user
    - √¥ `user_input` √¥ ƒë·ªÉ input c√¢u tr·∫£ l·ªùi c·ªßa user

    - Khi hi·ªÉn th·ªã "T·ªïng ƒëi·ªÉm c·ªßa b·∫°n l√† : ..." t·ª©c l√† ƒë√£ h·∫øt c√¢u h·ªèi, th√¨ ko hi·ªÉn th·ªã n√∫t `user_input` n·ªØa 
    - Ch·ªâ hi·ªÉn th·ªã n√∫t "Start" 1 l·∫ßn l√∫c ƒë·∫ßu. (    - ·∫®n n√∫t "Start" sau khi ƒë√£ hi·ªÉn th·ªã response_and_question)

````python
import pandas as pd
import gradio as gr

# Backend Function
def load_quiz_data(file_path, sheet_name):
    df = pd.read_excel(file_path, sheet_name=sheet_name)
    return df['question'].tolist(), df['checking_answer'].tolist(), df['explain_answer'].tolist()

def quiz_logic(questions, checking_answers, explain_answers, user_answer=None, current_question_index=0, total_score=0):
    if current_question_index >= len(questions) or pd.isna(questions[current_question_index]):
        return f"T·ªïng ƒëi·ªÉm c·ªßa b·∫°n l√†: {total_score}/{len(questions)}", gr.update(visible=False), gr.update(visible=False)
    
    if user_answer:
        if user_answer.strip().upper() == checking_answers[current_question_index]:
            assistant_response = f"ƒê√∫ng! Get 1 point.\n{explain_answers[current_question_index]}"
            total_score += 1
        else:
            assistant_response = f"Sai! Get 0 point.\n{explain_answers[current_question_index]}"
        
        response_and_question = (
            f"YOUR ANSWER: {user_answer}\n"
            f"FEEDBACK: \n{assistant_response}\n"
            f"---------------------"
        )
        
        current_question_index += 1
    else:
        response_and_question = ""
    
    if current_question_index < len(questions) and not pd.isna(questions[current_question_index]):
        response_and_question += f"\nC√¢u h·ªèi: {questions[current_question_index]}"
        return response_and_question, gr.update(visible=True), gr.update(visible=True), current_question_index, total_score
    else:
        response_and_question += f"\nT·ªïng ƒëi·ªÉm c·ªßa b·∫°n l√†: {total_score}/{len(questions)}"
        return response_and_question, gr.update(visible=True), gr.update(visible=False), current_question_index, total_score

# Frontend Function
def create_quiz_interface(file_path, sheet_name):
    questions, checking_answers, explain_answers = load_quiz_data(file_path, sheet_name)
    total_score = 0
    current_question_index = 0

    with gr.Blocks() as demo:
        start_button = gr.Button("Start")
        response_and_question = gr.Textbox(label="Response and Question", interactive=False, visible=False)
        user_input = gr.Textbox(label="Your Answer (A/B/C/D)", visible=False)
        
        def start_quiz():
            nonlocal current_question_index, total_score
            quiz_result, response_visibility, user_input_visibility, current_question_index, total_score = quiz_logic(
                questions, checking_answers, explain_answers, current_question_index=current_question_index, total_score=total_score
            )
            return quiz_result, gr.update(visible=False), response_visibility, user_input_visibility

        start_button.click(start_quiz, inputs=None, outputs=[response_and_question, start_button, response_and_question, user_input])
        
        def handle_user_input(user_answer):
            nonlocal current_question_index, total_score
            quiz_result, response_visibility, user_input_visibility, current_question_index, total_score = quiz_logic(
                questions, checking_answers, explain_answers, user_answer, current_question_index, total_score
            )
            return quiz_result, response_visibility, user_input_visibility

        user_input.submit(handle_user_input, inputs=user_input, outputs=[response_and_question, response_and_question, user_input])

    demo.launch()

# File path and sheet name
file_path = r'D:\OneDrive - Hanoi University of Science and Technology\ITE10-DS&AI-HUST\Learn&Task\PRODUCT_THECOACH\TASK5_GPTschatbot_TestingSale\Input\[SaleViewer]Define_b√†i_to√°n_SalerTesting.xlsx'
sheet_name = 'TN_P1P2'

# Create and launch the quiz interface
create_quiz_interface(file_path, sheet_name)
````

    - L·∫ßn l∆∞·ª£t load c√°c c√¢u h·ªèi t·ª´ c·ªôt `question` ƒë·ªÉ hi·ªán th·ªã ra (check xem n·∫øu d·ªØ li·ªáu l√† "None" t·ª©c l√† c√¢u h·ªèi ƒë√£ h·∫øt). Sau ƒë√≥ s·∫Ω l√† USER enter ƒë√°p √°n c·ªßa m√¨nh v√†o.
    V√≠ d·ª• 1 c√¢u h·ªèi:
    '''

    - run h√†m def scoring_short_essay_questions(QUESTION, CHECKING ANSWER) v·ªõi c√¢u tr·∫£ l·ªùi ƒë√∫ng ·ªü c·ªôt `checking_answer` =>

````python


from openai import OpenAI

def scoring_short_essay_questions(QUESTION, CHECKING_ANSWER, USER_ANSWER):
    client = OpenAI(api_key="")

    prompt = F"""
    
    
Act: You are an expert at scoring short essay questions. 
You are provided with 1 QUESTION and 1 CHECKING ANSWER for that QUESTION. After USER enters the answer, you will check USER ANSWER with CHECKING ANSWER. 

Instructions: 
- point: check USER ANSWER with CHECKING ANSWER to score on a 3-point scale 
0 points: The answer is completely unrelated to the question or has no content.
1 point: The answer may have a few relevant keywords but does not clarify the main idea at all (or lacks a connection).
2 points: The answer is correct but lacks analysis (or has analysis but is still sketchy).
3 points: The answer is correct, has clear analysis, and strong arguments. 

- comment: briefly in 2 sentences: the reason you gave such a score (in Vietnamese)
- answer_3_point: correct the user's original sentence to gain 3 points

Output: JSON format with 2 keys: `point` and `comment`and `answer_3_point`

QUESTION: 
{QUESTION}

CHECKING ANSWER: 
{CHECKING_ANSWER}
"""

    completion = client.chat.completions.create(
        model="gpt-4o-mini-2024-07-18",
        messages=[
            {"role": "system", "content": prompt},
            {"role": "user", "content": USER_ANSWER},
        
        ], 
        temperature = 0, 
        max_tokens=6000,
    )

    return completion.choices[0].message.content

question = "Theo b·∫°n, t·∫°i sao vi·ªác l·∫∑p ƒëi l·∫∑p l·∫°i l√† y·∫øu t·ªë quan tr·ªçng trong vi·ªác h·ªçc ngo·∫°i ng·ªØ? H√£y n√™u m·ªôt v√≠ d·ª• c·ª• th·ªÉ ƒë·ªÉ minh h·ªça. "
checking_answer = """
Answer: L·∫∑p ƒëi l·∫∑p l·∫°i gi√∫p c·ªßng c·ªë ki·∫øn th·ª©c v√† bi·∫øn n√≥ th√†nh ph·∫£n x·∫° t·ª± nhi√™n. V√≠ d·ª•, n·∫øu b·∫°n h·ªçc m·ªôt c·ª•m t·ª´ nh∆∞ng kh√¥ng l·∫∑p l·∫°i, b·∫°n s·∫Ω d·ªÖ qu√™n n√≥. Vi·ªác l·∫∑p l·∫°i gi√∫p ghi nh·ªõ s√¢u h∆°n v√† s·ª≠ d·ª•ng ƒë∆∞·ª£c ngay khi c·∫ßn. 

Tr√≠ch d·∫´n t·ª´ t√†i li·ªáu:  Luy·ªán t·∫≠p l·∫∑p l·∫°i: C·ªßng c·ªë ki·∫øn th·ª©c th√¥ng qua vi·ªác th·ª±c h√†nh l·∫∑p ƒëi l·∫∑p l·∫°i... Vi·ªác ghim s√¢u ƒë∆∞·ª£c c·ª•m v√†o ƒë·∫ßu ng∆∞·ªùi h·ªçc l√† m·ªôt ƒëi·ªÅu ti√™n quy·∫øt v√† c≈©ng l√† s·ª± kh√°c bi·ªát trong c√°ch thi·∫øt k·∫ø s·∫£n ph·∫©m.
"""

user_answer = "L·∫∑p l·∫°i nhi·ªÅu l·∫ßn gi√∫p c·ªßng c·ªë ki·∫øn th·ª©c v√† bi·∫øn n√≥ th√†nh ph·∫£n x·∫° t·ª± nhi√™n trong ng√¥n ng·ªØ. V√≠ d·ª•, n·∫øu b·∫°n th∆∞·ªùng xuy√™n luy·ªán t·∫≠p ph√°t √¢m m·ªôt t·ª´ m·ªõi, b·∫°n s·∫Ω d·∫ßn d·∫ßn nh·ªõ v√† s·ª≠ d·ª•ng t·ª´ ƒë√≥ m·ªôt c√°ch d·ªÖ d√†ng m√† kh√¥ng c·∫ßn ph·∫£i suy nghƒ© nhi·ªÅu."

response = scoring_short_essay_questions(question, checking_answer, user_answer)
print(response)
````

````python
import pandas as pd
import gradio as gr

# Backend Function
def load_quiz_data(file_path, sheet_name):
    df = pd.read_excel(file_path, sheet_name=sheet_name)
    try: 
        questions = df['question'].tolist()
        checking_answers = df['checking_answer'].tolist()
        explain_answers = df['explain_answer'].tolist()
    return questions, checking_answers, explain_answers

def quiz_logic(questions, checking_answers, explain_answers, user_answer=None, current_question_index=0, total_score=0):
    if current_question_index >= len(questions) or pd.isna(questions[current_question_index]):
        return f"T·ªïng ƒëi·ªÉm c·ªßa b·∫°n l√†: {total_score}/{len(questions)}", gr.update(visible=False), gr.update(visible=False)
    
    if user_answer:
        if user_answer.strip().upper() == checking_answers[current_question_index]:
            assistant_response = f"ƒê√∫ng! Get 1 point.\n{explain_answers[current_question_index]}"
            point = 1
            total_score += 1
        else:
            assistant_response = f"Sai! Get 0 point.\n{explain_answers[current_question_index]}"
            point = 0
        
        response_and_question = (
            f"YOUR ANSWER: {user_answer}\n"
            f"FEEDBACK: \n{assistant_response}\n"
            f"---------------------"
        )
        
        current_question_index += 1
    else:
        response_and_question = ""
    
    if current_question_index < len(questions) and not pd.isna(questions[current_question_index]):
        response_and_question += f"\nC√¢u h·ªèi: {questions[current_question_index]}"
        return response_and_question, gr.update(visible=True), gr.update(visible=True), current_question_index, total_score
    else:
        response_and_question += f"\nT·ªïng ƒëi·ªÉm c·ªßa b·∫°n l√†: {total_score}/{len(questions)}"
        return response_and_question, gr.update(visible=True), gr.update(visible=False), current_question_index, total_score

# Frontend Function
def create_quiz_interface(file_path, sheet_name):
    questions, checking_answers, explain_answers = load_quiz_data(file_path, sheet_name)
    total_score = 0
    current_question_index = 0

    with gr.Blocks() as demo:
        start_button = gr.Button("Start")
        response_and_question = gr.Textbox(label="Response and Question", interactive=False, visible=False)
        user_input = gr.Textbox(label="Your Answer (A/B/C/D)", visible=False)
        
        def start_quiz():
            nonlocal current_question_index, total_score
            quiz_result, response_visibility, user_input_visibility, current_question_index, total_score = quiz_logic(
                questions, checking_answers, explain_answers, current_question_index=current_question_index, total_score=total_score
            )
            return quiz_result, gr.update(visible=False), response_visibility, user_input_visibility

        start_button.click(start_quiz, inputs=None, outputs=[response_and_question, start_button, response_and_question, user_input])
        
        def handle_user_input(user_answer):
            nonlocal current_question_index, total_score
            quiz_result, response_visibility, user_input_visibility, current_question_index, total_score = quiz_logic(
                questions, checking_answers, explain_answers, user_answer, current_question_index, total_score
            )
            return quiz_result, response_visibility, user_input_visibility

        user_input.submit(handle_user_input, inputs=user_input, outputs=[response_and_question, response_and_question, user_input])

    demo.launch()

# File path and sheet name
file_path = r'D:\OneDrive - Hanoi University of Science and Technology\ITE10-DS&AI-HUST\Learn&Task\PRODUCT_THECOACH\TASK5_GPTschatbot_TestingSale\Input\[SaleViewer]Define_b√†i_to√°n_SalerTesting.xlsx'
sheet_name = 'TL_P1P2'

# Create and launch the quiz interface
create_quiz_interface(file_path, sheet_name)
````

````python
import pandas as pd
from openai import OpenAI
import gradio as gr
import json

def scoring_short_essay_questions(QUESTION, CHECKING_ANSWER, USER_ANSWER):
    client = OpenAI(api_key="")

    prompt = f"""
Act: You are an expert at scoring short essay questions. 
You are provided with 1 QUESTION and 1 CHECKING ANSWER for that QUESTION. After USER enters the answer, you will check USER ANSWER with CHECKING ANSWER. 

Instructions: 
- point: check USER ANSWER with CHECKING ANSWER to score on a 3-point scale 
0 points: The answer is completely unrelated to the question or has no content.
1 point: The answer may have a few relevant keywords but does not clarify the main idea at all (or lacks a connection).
2 points: The answer is correct but lacks analysis (or has analysis but is still sketchy).
3 points: The answer is correct, has clear analysis, and strong arguments. 

- comment: briefly in 2 sentences: the reason you gave such a score (in Vietnamese)
- answer_3_point: correct the user's original sentence to gain 3 points

Output: JSON format with 2 keys: `point`, `comment`, and `answer_3_point`

QUESTION: 
{QUESTION}

CHECKING ANSWER: 
{CHECKING_ANSWER}
"""

    completion = client.chat.completions.create(
        model="gpt-4o-mini-2024-07-18",
        messages=[
            {"role": "system", "content": prompt},
            {"role": "user", "content": USER_ANSWER},
        ], 
        temperature=0, 
        max_tokens=6000,
    )

    return completion.choices[0].message.content

def run_quiz(file_path, sheet_name):
    # Load the Excel file and the specific sheet
    df = pd.read_excel(file_path, sheet_name=sheet_name)

    # Extract relevant columns
    questions = df['question']
    checking_answers = df['checking_answer']

    # Initialize score
    total_score = 0
    current_question_index = 0

    def quiz_logic(user_answer=None):
        nonlocal current_question_index, total_score

        if current_question_index >= len(questions) or pd.isna(questions[current_question_index]):
            return f"T·ªïng ƒëi·ªÉm c·ªßa b·∫°n l√†: {total_score}/{len(questions)}", gr.update(visible=False), gr.update(visible=False)

        if user_answer:
            response = scoring_short_essay_questions(questions[current_question_index], checking_answers[current_question_index], user_answer)
            response_json = json.loads(response)
            point = response_json['point']
            comment = response_json['comment']
            answer_3_point = response_json['answer_3_point']

            total_score += point

            response_and_question = (
                f"- YOUR ANSWER: {user_answer}\n"
                f"- GET POINT: {point}/3 \n"
                f"- FEEDBACK: \n{comment}\n"
                f"- CORRECTED ANSWER: \n{answer_3_point}\n"
                f"---------------------"
            )

            current_question_index += 1
        else:
            response_and_question = ""

        if current_question_index < len(questions) and not pd.isna(questions[current_question_index]):
            response_and_question += f"\nC√¢u h·ªèi: {questions[current_question_index]}"
            return response_and_question, gr.update(visible=True), gr.update(visible=True)
        else:
            response_and_question += f"\nT·ªïng ƒëi·ªÉm c·ªßa b·∫°n l√†: {total_score}/{len(questions)*3}"  # M·ªói c√¢u 3 ƒëi·ªÉm
            return response_and_question, gr.update(visible=True), gr.update(visible=False)

    with gr.Blocks() as demo:
        start_button = gr.Button("Start")
        response_and_question = gr.Textbox(label="Response and Question", interactive=False, visible=False)
        user_input = gr.Textbox(label="Your Answer", visible=False)

        def start_quiz():
            quiz_result, response_visibility, user_input_visibility = quiz_logic()
            return quiz_result, gr.update(visible=False), response_visibility, user_input_visibility

        start_button.click(start_quiz, inputs=None, outputs=[response_and_question, start_button, response_and_question, user_input])

        def handle_user_input(user_answer):
            quiz_result, response_visibility, user_input_visibility = quiz_logic(user_answer)
            return quiz_result, response_visibility, user_input_visibility

        user_input.submit(handle_user_input, inputs=user_input, outputs=[response_and_question, response_and_question, user_input])

    demo.launch()

# Example usage
file_path = r'D:\OneDrive - Hanoi University of Science and Technology\ITE10-DS&AI-HUST\Learn&Task\PRODUCT_THECOACH\TASK5_GPTschatbot_TestingSale\Input\[SaleViewer]Define_b√†i_to√°n_SalerTesting.xlsx'
sheet_name = 'TL_P1P2'
run_quiz(file_path, sheet_name)
````
